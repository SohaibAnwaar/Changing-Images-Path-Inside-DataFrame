```python
import pandas as pd
df=pd.read_csv("diabetic.csv")
df.drop("Unnamed: 0",axis=1,inplace=True)
df.drop("Unnamed: 0.1",axis=1,inplace=True)
df.head()
```

```python
import os
df["image"]= df['Image_Path'].map(lambda x: x[30:])
df.head()
path="/home/imran/diabetic/resized/"

df['Image_Path'] = df['image'].map(lambda x: os.path.join(path+x))
df['exists'] = df['Image_Path'].map(os.path.exists)
df.drop("image",axis=1,inplace=True)
df.head()
```

```python
df.to_csv("14kDiabteic.csv")
```
